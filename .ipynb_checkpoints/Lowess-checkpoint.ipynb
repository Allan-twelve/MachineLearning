## 局部加权线性回归

```python
# 相关库引入
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from LinearRegression.Lowess import LWRegression
from DataSet_tools.data_split import train_test_split
from DataSet_tools.score import rSquare
%matplotlib inline
```

### 数据引入和分割（用的是鲍鱼年龄数据）

```python
data = pd.read_csv('Data/鲍鱼年龄.csv')
x = np.array(data.iloc[:, [1]])
y = np.array(data.iloc[:, [2]])
x_train, x_test, y_train, y_test = train_test_split(x, y, 0.2)
```

### 数据分布

```python
plt.plot(x, y, '.b')
```

### 模型拟合

```python
lwr = LWRegression(sigma=0.01)
lwr.fit(x_train, y_train)
```

### 模型预测评估

```python
y_pre = lwr.predict(x_test)
rSquare(y_test, y_pre)
```

```python
con = np.concatenate([x_test, y_pre], axis=1)
con.sort(axis=0)
plt.plot(x_train, y_train, '.b')
plt.plot(con[:, 0], con[:, 1], '-r')
```
