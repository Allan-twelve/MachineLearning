## 主成分分析

```python
# 数据引入，sklearn鸢尾花数据集和手写数据集
from sklearn.datasets import load_iris, load_digits
# 相关库引入
from DimensionalityReduction.PrincipalComponentAnalysis import PCA
from DataSet_tools.data_split import train_test_split
from DataSet_tools.score import accuracy
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
```

### 数据引入

```python
data = load_iris()
x = data.data
y = data.target.reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, 0.2)
```

### 降至二维

```python
pca = PCA(component=2)
x_train2D = pca.fit_transform(x_train)
x_test2D = pca.transform(x_test)
```

```python
plt.scatter(x_train2D[:, 0], x_train2D[:, 1], c=y_train)
```

### 降至三维

```python
pca = PCA(component=3)
x_train3D = pca.fit_transform(x_train)
x_test3D = pca.transform(x_test)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x_train3D[:, 0], x_train3D[:, 1], x_train3D[:, 2], c=y_train)
```

### 手写数字数据

```python
data = load_digits()
x = data.data
y = data.target.reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, 0.2)
```

### 降至三维

```python
pca = PCA(component=3)
x_train3D = pca.fit_transform(x_train)
x_test3D = pca.transform(x_test)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x_train3D[:, 0], x_train3D[:, 1], x_train3D[:, 2], c=y_train)
```

### 模型训练评估的对比

```python
from MultilayerPerceptron.BP import MLP
```

### 64维降至3维

```python
mlp = MLP(alpha=0.5, hidden=70)
mlp.fit(x_train3D, y_train)
y_pre = mlp.predict(x_test3D)
accuracy(y_test, y_pre)
```

### 原维度

```python
mlp = MLP(alpha=0.1, hidden=70)
mlp.fit(x_train, y_train)
y_pre = mlp.predict(x_test)
accuracy(y_test, y_pre)
```
