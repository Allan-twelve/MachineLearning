## 分割训练集和测试集

```python
dt = data[c.index].copy()
x = dt.drop(['SalePrice'], axis=1)
y = dt[['SalePrice']]
# 此处分割函数为随机分割，每次运行，训练集和测试集都不一样
x_train, y_train, x_test, y_test = train_test_split(x, y, 0.2)
```

### 引入梯度下降和正规方程法的线性模型

```python
G_lr = GradientDescent.LinearRegression(alpha=0.00000001, cycle=10000)
N_lr = NormalEquation.LinearRegression(normalization=True)
```

### 训练拟合

```python
G_lr.fit(x_train, y_train)
N_lr.fit(x_train, y_train)
G = G_lr.predict(x_test)
N = N_lr.predict(x_test)
```

```python
# 正规方程模型十次交叉验证, rSquare评价
cross_validation(x, y, N_lr, rSquare)
```

```python
# 模型评估
N1 = rSquare(y_test, N)
N2 = MAE(y_test, N)
N3 = MSE(y_test, N)
N4 = RMSE(y_test, N)
N5 = MAPE(y_test, N)
G1 = rSquare(y_test, G)
G2 = MAE(y_test, G)
G3 = MSE(y_test, G)
G4 = RMSE(y_test, G)
G5 = MAPE(y_test, G)
print('梯度下降：', 'r2:', G1, 'MAE:', G2, 'MSE:', G3, 'RMSE:', G4, 'MAPE:', G5)
print('正规方程：', 'r2:', N1, 'MAE:', N2, 'MSE:', N3, 'RMSE:', N4, 'MAPE:', N5)
```

## 对比sklearn

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
```

```python
sk_lr = LinearRegression()
sk_lr.fit(x_train, y_train)
sk_res = sk_lr.predict(x_test)
sk1 = r2_score(y_test, sk_res)
sk2 = mean_absolute_error(y_test, sk_res)
sk3 = mean_squared_error(y_test, sk_res)
sk4 = np.sqrt(mean_squared_error(y_test, sk_res))
sk5 = mean_absolute_percentage_error(y_test, sk_res)
print('sklearn：', 'r2:', sk1, 'MAE:', sk2, 'MSE:', sk3, 'RMSE:', sk4, 'MAPE:', sk5)
```

### 可以看出正规方程法和sklearn的模型各个评分基本一致， 但梯度下降就差很多，明显没有达到最优

## 数据归一化处理

```python
x_tr = pp.MinMaxScale(x_train)
x_te = pp.MinMaxScale(x_test, keep=True)
x_len = pp.len
x_min = pp.min
y_tr = pp.MinMaxScale(y_train)
y_te = pp.MinMaxScale(y_test, keep=True)
y_len = pp.len
y_min = pp.min
```

```python
G_lr_new = GradientDescent.LinearRegression(alpha=0.075, cycle=10000)
```

### 数据拟合

```python
G_lr_new.fit(x_tr, y_tr)
G_new = G_lr_new.predict(x_te)
```

```python
# 归一化模型评分
G_new = G_new * y_len[0] + y_min[0]
s1 = rSquare(y_test, G_new)
s2 = MAE(y_test, G_new)
s3 = MSE(y_test, G_new)
s4 = RMSE(y_test, G_new)
s5 = MAPE(y_test, G_new)
print('梯度下降(归一化)：', 'r2:', s1, 'MAE:', s2, 'MSE:', s3, 'RMSE:', s4, 'MAPE:', s5)
```

## 可以发现梯度下降和正规方程的最优解基本评分一致
